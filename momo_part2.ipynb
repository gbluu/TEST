```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

%matplotlib inline
```

```python
df = pd.read_csv(r'C:\Users\baolu\Desktop\raw.csv')
```

```python
df['UniqueID'] = df['User_id'].astype(str) + ' ' + df['Date'].astype(str)
df['Serviceid'] = df['Serviceid'].astype('str')
```

```python
df.drop_duplicates(inplace = True)
df.drop(['User_id', 'Date'], axis=1, inplace=True)
```

```python
df.set_index('UniqueID', inplace=True)
df
```

```python
encode_df = pd.get_dummies(df)
encode_df
```

```python
basket_sets = pd.pivot_table(encode_df, index='UniqueID', aggfunc='sum')
```

```python
itemsets = apriori(basket_sets, min_support=0.03, use_colnames=True)
rules = association_rules(itemsets, metric="lift", min_threshold=0.7)
```
