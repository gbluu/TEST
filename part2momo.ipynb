```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv(r'C:\Users\baolu\Desktop\result.csv')
```

```python
df["FirstServiceDate"]= pd.to_datetime(df["FirstServiceDate"])
df["SecondServiceDate"]= pd.to_datetime(df["SecondServiceDate"])
df["LastServiceDate"]= pd.to_datetime(df["LastServiceDate"])
```

```python
df['Cluster'] = df['LastServiceDate'] - df['FirstServiceDate']
df['Cluster'] = df['Cluster'].dt.days.apply(lambda x: 'Less30d' if x <=30 else 'More30d')
```
